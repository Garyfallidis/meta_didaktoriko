# Voxel Reconstruction Example

```python
from dipy.data import fetch_stanford_hardi, read_stanford_hardi
```

```python
fetch_stanford_hardi()
img, gtab = read_stanford_hardi()
```

```python
data = img.get_data()
print('data.shape (%d, %d, %d, %d)' % data.shape)
```

```python
%pylab inline
```

```python
imshow(data[:, :, 76/2, 0])
```

```python
from dipy.segment.mask import median_otsu
```

```python
data_masked, mask = median_otsu(data, 3, 2, True, range(0,10))
```

```python
data_masked.shape
```

```python
imshow(data_masked[:, :, 76/2, 0])
```

```python
print(gtab.bvals)
```

```python
len(gtab.bvals)
```

```python
from dipy.reconst.shm import CsaOdfModel
```

```python
csa_model = CsaOdfModel(gtab, 4)
```

```python
from dipy.data import get_sphere
sphere = get_sphere('symmetric724')
```

```python
data_small = data_masked[13:43, 44:74, 28:29]
```

```python
imshow(data_small[..., 0, 0].T, origin='lower')
```

```python
from dipy.viz import fvtk
ren = fvtk.ren()
```

```python
csa_fit = csa_model.fit(data_small)
```

```python
csa_fit.shm_coeff.shape
```

```python
csa_odfs = csa_fit.odf(sphere)
```

```python
import numpy as np
csa_odfs = np.clip(csa_odfs, 0, np.max(csa_odfs, -1)[..., None])
```

```python
spheres_actor = fvtk.sphere_funcs(csa_odfs, sphere, colormap='jet')
fvtk.add(ren, spheres_actor)
fvtk.show(ren)
```
