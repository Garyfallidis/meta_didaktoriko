### Direct multi-dimensional index operations with np.indices

```python
A = np.zeros((50, 50))
```

```python
grid = np.indices((10, 10))
grid
```

```python
rows, cols = grid
print(rows.shape)
print(cols.shape)
```

```python
A[rows, cols] = 1
imshow(A)
```

```python
A[rows+20, cols+20]=2
imshow(A)
```

```python
A = np.zeros((50, 50, 50))
rows, cols, depths = np.indices((10, 10, 10))
A[rows + 20, cols + 20, depths + 20] =  2
imshow(A[:, :, 25])
```
